```
import pandas as pd
df = pd.read_json('contacts.json')
###
asdf = df['Email']
aa = []
for i in asdf :
  if len(i) <= 5 :
    i = 'fail'
    aa.append(i)
  elif len(i) > 5 :
    aa.append(i)
##
qwer = df['Phone']
bb = []
for o in qwer :
  if len(o) <= 5 :
    o = 'fail'
    bb.append(o)
  elif len(o) > 5 :
    bb.append(o)
##
fasd = df['OrderId']
qq = []
for a in fasd :
  if len(a) <= 5 :
    a = 'fail'
    qq.append(a)
  elif len(a) > 5 :
    qq.append(a)
id = df['Id']

da = pd.DataFrame(aa,columns=["Email_N"])
db = pd.DataFrame(bb,columns=["Phone_N"])
dz = pd.DataFrame(qq,columns=["OrderId_N"])
result = pd.concat([id,da, db,dz], axis=1).reindex(da.index)
rs = result
rs
```
